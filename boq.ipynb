```python
import pandas as pd

df1 = pd.read_excel("boq_original.xlsx")
df2 = pd.read_excel("boq_updated.xlsx")
```

```python

# print(df1.head())
# print(df2.head())


merged = df1.merge(df2, on="item", how="outer", suffixes=("_old", "_new"))

# merged = pd.merge(df1,df2, left_index=True, right_index=True)

pd.set_option('display.max_columns', 1000)  # or 1000
pd.set_option('display.max_rows', 1000)  # or 1000
pd.set_option('display.max_colwidth', 199)  # or 199



print(merged)
```

```python
changed = merged [(merged["quantity (total)_old"] != merged["quantity (total)_new"]) & (~merged["price (total)_old"].isna()) & (~merged["price (total)_new"].isna())]

print(changed)
```

```python
added = merged[(merged["quantity (total)_old"].isna())]

print(added)
```

```python
removed = merged[(merged["quantity (total)_new"].isna())]

print (removed)
```

```python
changed['status'] = "changed"
added['status'] = "added"
removed['status'] = "removed"
```

```python
report = pd.concat([changed,added,removed])

report = report[["item" , "quantity (total)_old" , "quantity (total)_new", "status"]]

report.columns = ["item", "old_quantity", "new_quantity", "status"]

report = report.sort_values(by="status")

print(report)
```

```python
report.to_csv("report.csv", index=False)
```
